###1-2 N차원 배열의 데이터 타입

```python
#dtype 속성 없으면 자동으로 타입 유추
#타입 뒤의 숫자는 값 하나에 할당되는 bit 수

arr = np.array([1,2,3], dtype=float)
print(arr, arr.dtype)

arr = np.array([1.1, 2.2, 3.3], dtype=int)
print(arr, arr.dtype)

arr = np.array([0, 1, 2], dtype=bool)
print(arr, arr.dtype)

arr = np.array([0,1,2])
arr = arr.astype(np.float32)
print(arr, arr.dtype)
```

###3-6 벡터 연산의 장점

```python
import time
```

```python
arr = np.arange(99999999)
```

```python
#for
sum = 0
before = time.time()
for i in arr:
    sum += i
after = time.time()
print(sum, after-before)
```

```python
#vector
sum = 0
before = time.time()
sum = np.sum(arr)
after = time.time()
print(sum, after-before)
```

```python
arr1 = np.arange(99999999)
arr2 = np.arange(99999999)
```

```python
#for
dot = 0
before = time.time()
for i, j in zip(arr1, arr2):
    dot += i*j
after = time.time()
print(dot, after-before)
```

```python
#vector
before = time.time()
dot = np.dot(arr1, arr2)
after = time.time()
print(dot, after-before)
```

###5-1 배열의 형태 변경1: reshape

```python
arr = np.arange(12)
print(arr, arr.ndim)
```

```python
#reshape는 원본 변경 x
arr = arr.reshape(3,4) #원소의 개수를 맞춰줘야함
print(arr, arr.ndim)

arr = arr.reshape(4,3,1)
print(arr, arr.ndim)

arr = arr.reshape(4, -1) #-1을 넣으면 자동으로 사이즈 맞춰줌
print(arr, arr.ndim)
```

###5-2 배열의 형태 변경2: resize, ravel

```python
arr = np.arange(12)
print(arr)
```

```python
#resize는 원본 변경 o
arr.resize([3,4])
print(arr)
```

```python
#ravel: 1차원 배열로 변경
arr = arr.ravel()
print(arr)
```

###5-3 배열의 형태 변경3: expand_dims, squeeze

```python
arr = np.arange(12)
print(arr, arr.shape)

arr = np.expand_dims(arr, axis=0) #row 방향으로 확장, row 자리에 1 붙음
print(arr, arr.shape)
```

```python
arr = np.arange(12)
print(arr, arr.shape)

arr = np.expand_dims(arr, axis=1) #col 방향으로 확장, col 자리에 1 붙음
print(arr, arr.shape)
```

```python
arr = np.array([[1,2]])
print(arr, arr.shape, arr.ndim)

arr = np.squeeze(arr, axis=0) #size가 1인 axis만 squeeze 가능
print(arr, arr.shape, arr.ndim)
```

```python
arr = np.array([[[1],
                 [2],
                 [3]]])
print(arr, arr.shape, arr.ndim)

arr = np.squeeze(arr, axis=0)
print(arr, arr.shape, arr.ndim)
```

```python
arr = np.array([[[1,2,3]],
                [[4,5,6]]])
print(arr, arr.shape, arr.ndim)

arr = np.squeeze(arr, axis=1)
print(arr, arr.shape, arr.ndim)
```

```python
arr = np.array([[[3]],
                [[4]]])
print(arr, arr.shape, arr.ndim)

arr = np.squeeze(arr) #axis가 없으면 size가 1인 차원 모두 없애버림
print(arr, arr.shape, arr.ndim)
```
